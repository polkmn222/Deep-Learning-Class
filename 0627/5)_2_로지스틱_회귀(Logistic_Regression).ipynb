### * 시그모이드 함수(Sigmoid function)

image.png

```python
#  How to draw sigmoid function by using numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 그래프의 경우 w는1, b는 0임을 가정하면 아래와 같이 그릴 수 있습니다.

def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
X = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(X)


plt.plot(X,y,'g')
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```

```python
### 기울기의 변화에 따른 시그모이드 함수의 변화


y1 = sigmoid(0.5*X)
y2 = sigmoid(X)
y3 = sigmoid(2*X)

plt.plot(X,y1,'r', linestyle = '--') # w의 값이 0.5일 때
plt.plot(X,y2,'g')
plt.plot(X,y3,'b', linestyle = '--') # w의 값이 2일 때
plt.plot([0,0],[1.0,0.0],':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```

```python
### 값 덧셈 변화에 따른 시그모이드 함수의 변화

y1 = sigmoid(X+0.5)
y2 = sigmoid(X+1)
y3 = sigmoid(X+1.5)

plt.plot(X,y1,'r', linestyle = '--') # X + 0.5일 때
plt.plot(X,y2,'g')
plt.plot(X,y3,'b', linestyle = '--') # X + 1.5일 때
plt.plot([0,0],[1.0,0.0],':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```

### * 비용함수(Cost function)

로지스틱 회귀 또한 경사 하강법을 사용하여 가중치 w를 찾아내지만, 비용 함수로는 MSE를 사용하지 않습니다. 그 이유는 MSE를 로지스틱 회귀의 비용함수로 사용했을때는 좋지 않은 로컬 미니멈에 빠질 가능성이 지나치게 높아 문제 해결이 어려워집니다.

image.png

로지스틱 회귀는 MSE를 비용함수로 사용하면, SGD를사용하였을때 찾고자 하는 최소값이 아닌 잘못된 최소값에 빠질 가능성이 매우 높습니다. 이를 전체에 걸쳐 최소값인 ***글로벌 미니멈(Global Minimum)***이 아닌 특정 구역에서의 최소값인
***로컬 미니멈(Local Minimum)***에 도달했다라고 합니다.
